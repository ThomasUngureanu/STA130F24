Question 1:

The theoretical Simple Linear Regression model helps us understand the relationship between two variables called the predictor (independent) variable(X) and the outcome (dependent) variable (Y). In simple linear regression, the relationship between 
X and Y is assumed to be linear and can be expressed by the equation. Breaking down each component,

- Predictor Variable(X): This is the variable we use to predict or explain variations in Y. It’s usually observed data, assumed to be fixed or given.

- Outcome Variable(Y): This is the variable we’re trying to predict or model. In the real world, it’s affected by both the predictor variable and some random error.

- Intercept(β0): This is the expected value of Y when X is zero. It represents the starting point of the line on the Y axis. In practice, this value is determined from the data. 

- Slope (β1): This coefficient represents the change in Y for a one-unit change in X. It defines the steepness or incline of the relationship between X and Y. A positive slope suggests that as X increases, Y tends to increase, while a negative slope indicates the opposite. 

- Error Term (ϵ): The error term represents the deviation of observed data points from the line formed by β0+β1X. This deviation is assumed to be normally distributed with a mean of zero and a certain standard deviation. This term captures the randomness and variation in  Y that isn’t explained by X

Since the error term ϵ is normally distributed, Y is also normally distributed for a given value of X. In other words, for any given 
X value, Y follows a normal distribution centered around 
β0+β1X with some standard deviation. This randomness in Y means that even if we know X and the coefficients, there’s still an element of uncertainty in the prediction of Y.

Python Code Demonstration using numpy and scipy.stats
In this code:

We specify values for 
- β0, β1, and σ.

- We generate values for X as a sequence of numbers.

- We calculate Y based on the model Y= β0+β1X+ϵ, where ϵ is sampled from a normal distribution with mean 0 and standard deviation σ.

- Finally, we plot the observed data points and the true regression line without the error term for comparison


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Setting up the parameters for the linear model
np.random.seed(0)   # For reproducibility
beta_0 = 2          # Intercept
beta_1 = 0.5        # Slope
sigma = 1           # Standard deviation of the error term

# Generate predictor variable X
X = np.linspace(0, 10, 100)

# Generate the outcome variable Y with normally distributed error term
epsilon = norm.rvs(loc=0, scale=sigma, size=X.shape)  # Random error term with mean=0 and std=sigma
Y = beta_0 + beta_1 * X + epsilon                     # Linear relationship plus error

# Plotting the generated data
plt.figure(figsize=(10, 6))
plt.scatter(X, Y, color="blue", alpha=0.5, label="Observed Data")
plt.plot(X, beta_0 + beta_1 * X, color="red", label="True Regression Line (without error)")
plt.xlabel("Predictor Variable X")
plt.ylabel("Outcome Variable Y")
plt.title("Simple Linear Regression Model with Normally Distributed Error")
plt.legend()
plt.show()
```

How fitted_model.fittedvalues are calculated:
1. Coefficients (Intercept and Slope): After fitting the model, the fitted_model.params object contains the estimated coefficients for the intercept and the slope. For a simple linear regression of the form

- fitted_model.params["Intercept"] (or fitted_model.params[0]) gives the estimated intercept β0.

- fitted_model.params["X"] (or fitted_model.params[1]) gives the estimated slope β1^β 1.

2. Deriving Fitted Values Using the Coefficients: Once we have the estimated coefficients β0 and β1, we can calculate the fitted values for Y for each X value using the equation Y=β0+β1X. Here, Y is the fitted value, calculated by plugging in each observed X value from the dataset into the equation along with the estimated intercept and slope.

3. Automatic Calculation in fitted_model.fittedvalues: The attribute fitted_model.fittedvalues performs this calculation for each data point automatically, applying the equation Y=β0+β1X to each X in the dataset using the values from fitted_model.params.

- The result of np.allclose(model.fittedvalues, fitted_values_manual) confirms that the manually calculated fitted values match fitted_model.fittedvalues, proving that they’re derived directly from the model’s intercept and slope estimates.

```python
import numpy as np
import pandas as pd
import statsmodels.formula.api as smf

# Simulating data as before
np.random.seed(0)
X = np.linspace(0, 10, 100)
beta_0 = 2
beta_1 = 0.5
epsilon = np.random.normal(0, 1, size=X.shape)
Y = beta_0 + beta_1 * X + epsilon

# Create a DataFrame and fit the model
data = pd.DataFrame({"X": X, "Y": Y})
model = smf.ols("Y ~ X", data=data).fit()

# Coefficients from the model
intercept = model.params["Intercept"]
slope = model.params["X"]

# Calculate fitted values manually
fitted_values_manual = intercept + slope * data["X"]

# Display comparison
print("Model Fitted Values (from fitted_model.fittedvalues):")
print(model.fittedvalues.head())
print("\nManually Calculated Fitted Values:")
print(fitted_values_manual.head())

# Check if they are nearly identical
print("\nAre the values identical? ", np.allclose(model.fittedvalues, fitted_values_manual))
```

The line chosen for the fitted model in Ordinary Least Squares (OLS) regression is the one that minimizes the sum of the squared differences (or "residuals") between the observed data points and the line itself. This line represents the "best fit" because it’s the one where the total squared distance between the actual 
Y values and the predicted Y values (from the line) is the smallest. The "squares" are used because squaring each residual:
1. Eliminates negative values, treating both under- and over-predictions equally.
2. Emphasizes larger errors more than smaller ones, which helps the model focus on reducing larger deviations from the line, improving overall accuracy.

Question 6:
1. Proportion of Variation Explained by the Model
The expression for the proportion of variation in 
Y explained by the model is often represented by:
r-squared= 1−SSR/SST
where:
- SSR (Sum of Squared Residuals) is the sum of squared differences between observed Y values and the fitted (predicted) Y values by the model. It represents the variation in not explained by the model.
- SST (Total Sum of Squares) is the sum of squared differences between observed Y values and the mean of Y. It represents the total variation in Y without accounting for any model. By subtracting the ratio SSR/SST from 1, we get the proportion of Y's total variation that is explained by the model’s predictions. Thus, this expression measures the degree to which our model accounts for the variability in Y.

2. Interpretation of fitted_model.rsquared
fitted_model.rsquared in statsmodels directly calculates this proportion of explained variation, also known as R-squared. Because it reflects how well the model explains the variation in 
Y, it’s commonly interpreted as a measure of the model's accuracy. An R^2 value closer to 1 indicates that the model captures a larger portion of the variability in Y, while a value closer to 0 suggests the model has little explanatory power.

3. The Role of 
np.corrcoef(...)[0,1]^2 in Simple Linear Regression
- In the context of Simple Linear Regression, the squared correlation coefficient \text{np.corrcoef(Y, Y_{\text{pred}})[0,1]}^2 measures the strength of the linear relationship between the observed valuesY and the predicted values Y pred (from fitted_model.fittedvalues). This squared correlation coefficient is mathematically equivalent to R2 in Simple Linear Regression.

(A) Permutation Test
In a permutation test, we test the null hypothesis that the mean eruption duration is the same for both "short" and "long" wait times. The key steps are:
1. Calculate the Observed Difference in Means: Compute the actual difference in mean eruption durations between "short" and "long" groups.
2. Shuffle Labels: Combine the data from both groups, then randomly shuffle the labels ("short" and "long") to simulate a scenario where there is no actual difference between the groups.
3. Recalculate Mean Difference: Split the shuffled data back into two groups and calculate the difference in means.
4. Repeat and Assess: Repeat the shuffling many times (e.g., 10,000 times) to build a distribution of mean differences under the null hypothesis. The p-value is then calculated as the proportion of shuffled mean differences as extreme as or more extreme than the observed difference. 

```python
import numpy as np
import pandas as pd
import seaborn as sns

# Load and prepare the dataset
geyser_data = sns.load_dataset('geyser').dropna(subset=['waiting', 'duration'])
geyser_data['katagory'] = ['short' if w < 68 else 'long' for w in geyser_data['waiting']]

# Observed difference in means
mean_diff_observed = geyser_data[geyser_data['katagory'] == 'long']['duration'].mean() - \
                     geyser_data[geyser_data['katagory'] == 'short']['duration'].mean()

# Permutation test
n_permutations = 10000
permuted_diffs = []

for _ in range(n_permutations):
    # Shuffle the 'katagory' labels
    shuffled_katagory = np.random.permutation(geyser_data['katagory'])
    # Assign the shuffled labels to the dataset
    geyser_data['shuffled_katagory'] = shuffled_katagory
    # Calculate mean difference with shuffled labels
    permuted_diff = geyser_data[geyser_data['shuffled_katagory'] == 'long']['duration'].mean() - \
                    geyser_data[geyser_data['shuffled_katagory'] == 'short']['duration'].mean()
    permuted_diffs.append(permuted_diff)

# Calculate p-value
p_value = np.mean(np.abs(permuted_diffs) >= np.abs(mean_diff_observed))
print("Permutation Test p-value:", p_value)
```

(B) Bootstrap Confidence Interval
For the bootstrap confidence interval, we aim to estimate a 95% confidence interval for the difference in mean eruption durations between "short" and "long" wait times. Here’s the process:

Bootstrap Sampling within Groups: For each bootstrap iteration, draw random samples with replacement from each group separately (short and long wait times).
Calculate Difference in Means: Compute the difference in means between the bootstrapped samples of the "short" and "long" groups.
Repeat and Quantile Calculation: Repeat the resampling process many times (e.g., 10,000 times) and calculate the 2.5th and 97.5th percentiles of the differences to get a 95% confidence interval.

```python
# Bootstrap confidence interval
n_bootstrap = 10000
bootstrapped_mean_diffs = []

for _ in range(n_bootstrap):
    # Sample with replacement within each group
    sample_short = geyser_data[geyser_data['katagory'] == 'short']['duration'].sample(frac=1, replace=True)
    sample_long = geyser_data[geyser_data['katagory'] == 'long']['duration'].sample(frac=1, replace=True)
    # Calculate the difference in means
    bootstrapped_diff = sample_long.mean() - sample_short.mean()
    bootstrapped_mean_diffs.append(bootstrapped_diff)

# Calculate 95% confidence interval
ci_lower, ci_upper = np.quantile(bootstrapped_mean_diffs, [0.025, 0.975])
print("95% Bootstrap Confidence Interval for Difference in Means:", (ci_lower, ci_upper))
```

(a) Explanation of Sampling Approaches
- Permutation Test: By shuffling labels, we simulate a scenario where there is no difference between groups. This process creates a distribution of mean differences under the null hypothesis, allowing us to assess whether the observed difference in means is likely to occur by random chance.
- Bootstrap Confidence Interval: This method resamples the original data within each group independently, maintaining the group structure. By repeatedly calculating the difference in means for these resampled datasets, we create a distribution of mean differences. The confidence interval provides a range of plausible values for the true difference in means, assuming that the observed sample reflects the population.

- Permutation Test: This is a non-parametric method that does not assume normality. By shuffling labels, we create a sampling distribution under the null hypothesis, making it robust to deviations from normality and ideal for small or skewed datasets.
- Bootstrap Confidence Interval: Like the permutation test, bootstrapping is non-parametric and does not assume normality. It estimates the variability of the difference in means by resampling within each group, making it more flexible in terms of distributional assumptions.
